```python
import pandas as pd

from os import listdir
from os.path import isfile, join
from sklearn.tree import DecisionTreeRegressor
```

## 1. Data

```python
path = "./data/"

files_dict = {}
for file in listdir(path):
    files_dict[file] = pd.read_csv(path + file)
    if file == "train.csv":
        files_dict[file] = files_dict[file].drop("Cabin", axis = 1)
    files_dict[file] = files_dict[file].dropna()
    display(files_dict[file].head())
```

## 2. Model

```python
titanic_model = DecisionTreeRegressor()
y = files_dict["train.csv"].Survived
X = files_dict["train.csv"][
    [col for col in files_dict["train.csv"].columns
     if col not in ["Survived", "PassengerId", "Name","SibSp", "Parch", "Cabin", "Ticket"]]
]
```

## 3. Processing the data

```python
X = pd.get_dummies(X)
```

```python
titanic_model.fit(X, y)

print("First in-sample predictions:", titanic_model.predict(X.head()))
print("Survived:", y.head().tolist())

print(titanic_model.predict(X))
print(y)
```

```python
y_pred = titanic_model.predict(X)
```
